```python
import numpy as np
```

```python
#Vector Addition
vector1 = np.array([5, 10, 15])
vector2 = np.array([1, 8, 21])

result1 = vector1 + vector2
print(result1)
```

```python
#Vector Multiplication
vector3 = np.array([[15, 30, 16], [25, 50, 27], [12, 18, 40]])
vector4 = np.array([[1, 16, 25], [88, 76, 4], [18, 36, 77]])

result2 = np.matmul(vector3, vector4)
print(result2)
```
